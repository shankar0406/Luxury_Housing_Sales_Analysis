```python
# Function to clean currency text and extract numeric value
def clean_ticket_price(value):
    if pd.isna(value):
        return None
    # Remove anything that’s not a digit, decimal point, or minus sign
    cleaned = re.sub(r'[^0-9.]', '', str(value))
    try:
        return float(cleaned)
    except ValueError:
        return None  # In case conversion fails
sales_annalysis['Ticket_Price_Cr'] = sales_annalysis['Ticket_Price_Cr'].apply(clean_ticket_price)
```

```python
sales_annalysis['Ticket_Price_Cr']=sales_annalysis['Ticket_Price_Cr'].astype('float64')
sales_annalysis['Purchase_Quarter']=sales_annalysis['Purchase_Quarter'].astype('datetime64[ns]')
```

```python
sales_annalysis['Unit_Size_Sqft']=sales_annalysis['Unit_Size_Sqft'].replace(-1.0,np.nan)
```

```python
sales_annalysis.iloc[145]
```

```python
print("Missing before:", sales_annalysis['Unit_Size_Sqft'].isna().sum())
```

```python


# Impute Unit_Size_Sqft using mean by Configuration & Micro_Market
sales_annalysis['Unit_Size_Sqft'] = sales_annalysis.groupby(['Configuration', 'Micro_Market'])['Unit_Size_Sqft'].transform(lambda x: x.fillna(x.mean()))
```

```python
sales_annalysis.iloc[145]
```

```python
sales_annalysis.info()
```

Continuous variables like Unit_Size_Sqft can have many unique values.
To group them meaningfully, you should bin them into ranges.

```python
sales_annalysis.Ticket_Price_Cr.isnull().sum()
```

```python
# Create binned Unit_Size_Sqft for grouping
sales_annalysis['Size_Bin'] = pd.cut(sales_annalysis['Unit_Size_Sqft'], bins=10)

# Impute missing Ticket_Price_Cr using mean by (Configuration + Micro_Market + Size_Bin)
sales_annalysis['Ticket_Price_Cr'] = sales_annalysis.groupby(
    ['Configuration', 'Micro_Market', 'Size_Bin']
)['Ticket_Price_Cr'].transform(lambda x: x.fillna(x.mean()))


# Drop helper bin column
sales_annalysis.drop(columns=['Size_Bin'], inplace=True)
```

```python
sales_annalysis.info()
```

```python
sales_annalysis.Amenity_Score.isnull().sum()        
```

```python
# Step 2: Create price bins (to represent luxury levels)
sales_annalysis['Price_Bin'] = pd.cut(sales_annalysis['Ticket_Price_Cr'], bins=10)

# Step 3: Impute missing Amenity_Score using mean grouped by Configuration, Micro_Market, and Price_Bin
sales_annalysis['Amenity_Score'] = sales_annalysis.groupby(['Configuration', 'Micro_Market', 'Price_Bin'])['Amenity_Score'].transform(lambda x: x.fillna(x.mean()))

sales_annalysis.drop(columns=['Price_Bin'], inplace=True)
```

```python
def get_booking_flag(row):

    if str(row['Possession_Status']).strip().lower() in ['ready to move','completed']:
        return 1  # booked / ready property
    else:
        return 0  # Not booked or still under development

sales_annalysis['Booking_Flag_edits'] = sales_annalysis.apply(get_booking_flag, axis=1)
```
